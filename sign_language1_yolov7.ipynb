Preparing image path in txt file

```python
import os
```

```python
train_image_path = "/content/images/train"
val_image_path =  "/content/images/val"
```

```python
# setting paths of training images in train.txt
with open('train.txt', "a+") as f:
    image_list = os.listdir(train_image_path)
    for img in image_list:
        f.write(os.path.join(train_image_path,img+'\n'))
    f.close()

# setting paths of validation images in val.txt

with open('val.txt', "a+") as f:
    image_list = os.listdir(val_image_path)
    for img in image_list:
        f.write(os.path.join(val_image_path, img+'\n'))
    print('Done')
    f.close()
```
